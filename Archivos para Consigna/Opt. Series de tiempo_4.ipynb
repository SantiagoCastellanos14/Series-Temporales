# Estacionaridad

Una serie temporal estacionaria es aquella cuyas propiedades estadsticas, como la media, la varianza y la autocovarianza, son todas constantes y no una funcin del tiempo.

Si podemos suponer que estas propiedades estadsticas no cambiarn con el tiempo, entonces podemos predecir que el futuro de la serie temporal tendr las mismas propiedades estadsticas o proporcionales.

Como resultado, la mayora de los mtodos estadsticos de prediccin se basan en la suposicin de que la serie temporal es aproximadamente estacionaria, pero la mayora de las series temporales que veremos no lo son.

La no estacionaridad pueden provenir de cambios deterministas como la tendencia o las fluctuaciones estacionales, o porque las propiedades estocsticas del proceso tienen una raz unitaria. En el primer caso podemos eliminar el componente determinstico, y en el segundo podemos probar la presencia de la raz unitaria y diferenciarla.

En esta notebook explorar algunas de las principales formas en que podemos descomponer, probar y procesar nuestra serie de tiempo. Encontrar la secuencia de transformaciones necesarias para estacionarizar una serie de tiempo a menudo brinda pistas importantes en la bsqueda de un modelo de pronstico apropiado.


<a id='unit root'></a>
### Unit Root

Las races unitarias son una de las causas de la no estacionariedad y es importante entender porqu. Un proceso estocstico lineal tiene una raz unitaria si el valor de la raz de la ecuacin caracterstica del proceso es igual a 1. En una serie de tiempo, una raz unitaria provoca una tendencia estocstica, a veces denominada caminata aleatoria con deriva.

Presentaremos dos test de estacionaridad muy usados.



<a id='adf'></a>
#### Augmented Dickey-Fuller

La hiptesis nula para esta prueba es que existe una raz unitaria. La hiptesis alternativa es que la serie temporal es estacionaria o estacionaria en tendencia. Si el estadstico de prueba es menor que el valor crtico, podemos rechazar la hiptesis nula y decir que la serie es estacionaria. Adems, si el valor de p es menor que alfa, podemos rechazar la hiptesis nula.
<a id='kpss'></a>



#### KPSS


KPSS determina si una serie temporal es estacionaria en torno a una tendencia media o lineal. La hiptesis nula es que los datos son estacionarios, la hiptesis alternativa para la prueba es que los datos no son estacionarios.

Para interpretar, si el estadstico de prueba es mayor que los valores crticos, entonces se rechaza la hiptesis nula y la serie es no estacionaria. Tambin podemos usar el valor p devuelto y compararlo con el nivel alfa, donde si el valor p es menor que nuestro alfa, rechazamos la hiptesis nula.

Si no rechazamos la hiptesis nula, significa que nuestra serie de tiempo es estacionaria o estacionaria en tendencia, porque KPSS clasifica una serie como estacionaria en ausencia de una raz unitaria.

Debido a que hay varios tipos de estacionaridad, podemos usar KPSS y ADF en conjunto para determinar qu transformaciones hacer. Si KPSS = Estacionario y ADF = No estacionario, nuestra serie de tiempo es estacionaria en tendencia y debemos eliminar la tendencia para que sea estrictamente estacionaria. Si KPSS = No estacionario y ADF = Estacionario, entonces nuestra serie de tiempo es estacionaria en diferencia y necesitamos diferenciar nuestra serie.

## Hacia la estacionaridad

Ahora que sabemos qu es estacionario y cmo probarlo, as es como podemos crear una serie estacionaria, salvo componentes estacionales y deterministas fuertes, que veremos ms adelante. Dos prcticas comunes son la transformacin y la diferenciacin de Box-Cox.

<a id='box-cox'></a>
#### Box-Cox Transformation

Si nuestra serie de tiempo tiene una tendencia cuadrtica subyacente, podramos tomar la raz cuadrada y hacerla lineal, acercando los datos un paso ms a lo estacionario, para eso sirve una transformacin de Box-Cox. Una transformacin de Box-Cox intenta encontrar la transformacin de potencia apropiada para hacer que nuestros datos sean ms lineales y ms normalmente distribuidos. Si nuestros datos parecen tener una tendencia de potencia, entonces son buenos candidatos para una transformacin de Box-Cox. Podemos pasar a Box-Cox un argumento para lambda, el parmetro de transformacin, para realizar una transformacin de potencia o podemos pasar ningn argumento y automticamente estimar y devolver un valor lambda.
* lambda = -1. es una transformada recproca.
* lambda = -0.5 es una transformada de raz cuadrada recproca.
* lambda = 0.0 es una transformacin logartmica.
* lambda = 0.5 es una transformacin de raz cuadrada.
* lambda = 1.0 no es ninguna transformacin.

Podemos probar la normalidad de varias maneras:
*Shapiro-Wilk
* Jarque-Bera: comprueba la curtosis y la asimetra frente a una distribucin normal
* Grfica Q-Q (Cuantil-Cuantil): Se puede usar una grfica de los cuantiles de dos distribuciones
La normalidad no es necesaria para la estacionariedad pero lo es para muchos otros modelos y tcnicas.


```python
def adf_test(timeseries):
    print ('Results of Dickey-Fuller Test:')
    print('Null Hypothesis: Unit Root Present')
    print('Test Statistic < Critical Value => Reject Null')
    print('P-Value =< Alpha(.05) => Reject Null\n')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput[f'Critical Value {key}'] = value
    print (dfoutput, '\n')
```

```python

def kpss_test(timeseries, regression='c'):
    # Whether stationary around constant 'c' or trend 'ct
    print ('Results of KPSS Test:')
    print('Null Hypothesis: Data is Stationary/Trend Stationary')
    print('Test Statistic > Critical Value => Reject Null')
    print('P-Value =< Alpha(.05) => Reject Null\n')
    kpsstest = kpss(timeseries, regression=regression)
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
    for key,value in kpsstest[3].items():
        kpss_output[f'Critical Value {key}'] = value
    print (kpss_output, '\n')
```

Normalmente, la cantidad correcta de diferenciacin es el orden ms bajo de diferenciacin que produce una serie de tiempo que flucta alrededor de un valor medio bien definido y cuyo grfico de funcin de autocorrelacin (ACF) decae con bastante rapidez a cero. ### Descomposicin en estacionalidad



La descomposicin de la estacionalidad es una herramienta de anlisis que nos proporciona un marco para pensar sobre los diferentes componentes de nuestra serie temporal. Podemos descomponer nuestras series para entender la influencia de cada componente y usar eso para guiar nuestra preparacin de datos, seleccin de modelos y ajuste de modelos. Como vimos en la seccin sobre suavizacin exponencial, modelos que consideran la estacionalidad frente a los que no tienen un efecto prominente en nuestra capacidad de predecir la serie.

Podemos pensar en nuestra serie temporal como compuesta por una combinacin de nivel, tendencia, estacionalidad y ruido.

- nivel: El valor promedio de la serie
- tendencia: El valor creciente o decreciente en la serie
- estacionalidad: El ciclo de corto plazo que se repite en la serie
- ruido: La variacin aleatoria en la serie, denominada residual por la funcin de descomposicin_estacional




Una estacin es un perodo de tiempo fijo que contiene la repeticin completa del patrn de su serie temporal. Preste mucha atencin, sus datos pueden repetirse todos los das, pero si los fines de semana se ven diferentes a los das de la semana, entonces su temporada es una semana, no un da.

La descomposicin de la estacionalidad es una herramienta de anlisis que nos proporciona un marco sobre cmo pensar en los diferentes componentes de nuestra serie temporal. Podemos descomponer nuestra serie para comprender la influencia de cada componente y usar eso para guiar nuestra preparacin de datos, seleccin de modelos y ajuste de modelos. Como vimos en la seccin sobre suavizado exponencial, los modelos que consideran la estacionalidad frente a los que no tienen un efecto destacado en nuestra capacidad para pronosticar la serie.



La descomposicin estacional en `statsmodels` solo requiere que indiquemos si el modelo es aditivo o multiplicativo.

Esta es una descomposicin ingenua, utilizada principalmente para la descripcin de datos.

El modelo aditivo es Y[t] = T[t] + S[t] + e[t]

El modelo multiplicativo es Y[t] = T[t] * S[t] * e[t]

Los resultados se obtienen estimando primero la tendencia aplicando un filtro de convolucin a los datos. Luego, la tendencia se elimina de la serie y el promedio de esta serie sin tendencia para cada perodo es el componente estacional devuelto.




### Additive vs. Multiplicative

En una serie de tiempo multiplicativa, los componentes descompuestos se multiplican para formar la serie de tiempo. En una serie multiplicativa hay tendencia creciente, aumenta la amplitud de la actividad estacional y todo se vuelve ms exagerado. La tendencia multiplicativa se parece ms a una curva exponencial y la estacionalidad multiplicativa tiene ondas que crecen en amplitud a lo largo del tiempo.

En un modelo aditivo asumimos que los componentes de la serie de tiempo tienen un efecto aditivo, que la amplitud del efecto estacional es aproximadamente la misma, que el tamao de los residuos es mayormente constante.

Los datos del mundo real no siempre sern puramente aditivos o multiplicativos, puede haber secciones de cualquier tipo. Es importante probar varios parmetros y ver cul se ajusta mejor a nuestros datos.

```python
decomposition_df = pd.DataFrame(monthly_df.O_3)
seasonal_a = seasonal_decompose(decomposition_df, model='additive')
seasonal_m = seasonal_decompose(decomposition_df, model='multiplicative')
fig_1 = seasonal_a.plot()
fig_2 = seasonal_m.plot()
fig_1.suptitle('Additive Seasonal Decomposition', fontsize=25)
fig_1.set_figheight(10)
fig_1.set_figwidth(20)
fig_2.suptitle('Multiplicative Seasonal Decomposition', fontsize=25)
fig_2.set_figheight(10)
fig_2.set_figwidth(20)
plt.show()
```

# Suavizado

El suavizado es una tcnica que podemos utilizar para reducir el ruido en nuestra serie temporal y realizar pronsticos. El pronstico con tcnicas de suavizado como el suavizado exponencial lineal (LES) y el suavizado exponencial simple (SES) puede funcionar para datos no estacionales donde la serie temporal es estacionaria localmente y tiene una tendencia leve.





Para series de tiempo ms complejas, existen tcnicas como los mtodos estacionales de Holt-Winters que se basan en tcnicas de suavizado simples. No profundizaremos mucho en los mtodos de pronstico de Holt-Winters, pero los presentaremos, presentaremos cmo se relacionan con tcnicas ms simples y, con suerte, entenderemos cmo pensar en nuestra serie de tiempo como una combinacin de componentes y cmo el suavizado afecta la distribucin subyacente. de nuestros datos.



Las tcnicas de suavizado son fundamentales para las series temporales y se utilizan en varias aplicaciones, incluida la familia ARMA de modelos de pronstico. Al suavizar, debemos prestar atencin a la varianza y la forma de nuestras distribuciones de datos, ya que los modelos excesivamente suavizados pueden mostrar un alto sesgo, y el suavizado puede reducir demasiada varianza segn nuestros parmetros.



<a id='sma'></a>
### Simple Moving Average (SMA)
El promedio mvil simple es un promedio igualmente ponderado de n trminos de datos. El promedio mvil simple es excelente cuando faltan datos o son muy irregulares. Es primitivo pero robusto. Nuestros datos de contaminantes son increblemente ruidosos, particularmente en su escala horaria original. SMA es una buena herramienta de preprocesamiento para suavizar gradualmente nuestros datos al nivel que queremos.

```python
SMA_df= pd.DataFrame(monthly_df.NO_2)
for n in range(1, 6):
    label = 'window_'+str(n*3)
    SMA_df[label] = SMA_df['NO_2'].rolling(window=n*3).mean()

SMA_df.plot(figsize=(20, 5), title='NO_2 Moving Averages')
plt.show()
```

```python
error = {}
for col in SMA_df.columns[1:]:
    series = SMA_df[['NO_2', col]]
    series = series.dropna(axis=0)
    error[col] = mean_squared_error(series['NO_2'], series[col])
    print(f'Moving Average {col} MSE: ', round(error[col], 2))

metadata = {'mean':[], 'variance':[], 'skew':[], 'kurtosis':[]}
for col in SMA_df:
    metadata['mean'].append(SMA_df[col].mean())
    metadata['variance'].append(SMA_df[col].var())
    metadata['skew'].append(SMA_df[col].skew())
    metadata['kurtosis'].append(SMA_df[col].kurt())
```

```python

plt.figure(figsize=(10,10))
plt.suptitle('Changes in Series Distribution as Window Size Increases')
plt.subplot(221)
plt.title('NO2 Mean')
plt.plot(metadata['mean'])
plt.subplot(222)
plt.title('NO2 Variance')
plt.plot(metadata['variance'])
plt.subplot(223)
plt.title('NO2 Skewness')
plt.plot(metadata['skew'])
plt.subplot(224)
plt.title('NO2 Kurtosis')
plt.plot(metadata['kurtosis'])
plt.show()
```

### Simple Exponential Smoothing (SES)
El suavizado exponencial es un promedio ponderado de n trminos de datos pasados, donde los pesos decaen exponencialmente. Podemos usar el suavizado exponencial simple cuando los datos no son estacionales y muestran una media variable en el tiempo sin una tendencia constante. SES es lo mismo que un modelo ARIMA(0, 1, 1) sin una constante estacional. Proporcionamos a SES un valor, alfa, un parmetro de suavizado que controla la velocidad a la que los pesos decaen exponencialmente. Los valores grandes significan que el modelo presta mucha atencin a las observaciones pasadas ms recientes, y un alfa ms pequeo considera una historia mayor.



### Double Exponential Smoothing

El suavizado exponencial doble es para analizar datos que muestran una tendencia, agrega ecuaciones adicionales al suavizado exponencial simple para capturar el nivel y la tendencia de la serie, utilizando el parmetro beta. El suavizado exponencial doble tambin se denomina mtodo de tendencia lineal de Holt y en el paquete statsmodels usamos `ExponentialSmoothing` para implementar el suavizado exponencial doble y triple.


Al usar el paquete statsmodels todo lo que tenemos que indicar es si la tendencia es aditiva o multiplicativa. Esto se discute ms adelante en el ncleo, pero simplemente, la tendencia aditiva y multiplicativa se trata de si la tendencia es consistente o parece crecer, como una funcin exponencial, por ejemplo.


Si la tendencia es consistente entonces es aditiva, si la tendencia crece entonces es multiplicativa. Tambin parte de la implementacin de suavizado exponencial de statsmodels es un argumento `amortiguado`. El mtodo lineal de Holt tiende a mostrar una tendencia constante indefinidamente en el futuro.

 El parmetro `amortiguado` corrige esta tendencia infinita amortiguando la tendencia. Hay un parmetro amortiguado entre 0 y 1 donde los valores ms cercanos a 1 son los mismos que en un modelo no amortiguado. Podemos establecer estos parmetros nosotros mismos, pero en la prctica, simplemente establecemos si incluir o no el parmetro amortiguado.

### Triple Exponential Smoothing

Si sus datos muestran tendencia y estacionalidad, el suavizado exponencial triple se basa en las ecuaciones utilizadas en simple y doble y agrega una tercera ecuacin para manejar la estacionalidad. Tambin llamado Holt-Winters Exponential Smoothing, un nuevo parmetro, gamma, se utiliza para influir en la estacionalidad. Adems, indicamos si la estacionalidad es aditiva o multiplicativa como hicimos con la tendencia.

```python
smoothing_df = pd.DataFrame(monthly_df.O_3)
smoothing_df.plot(figsize=(20, 5), title='Original Monthly O_3 Time Series')
plt.show()
es_simple = SimpleExpSmoothing(smoothing_df.O_3).fit()
es_double = ExponentialSmoothing(smoothing_df.O_3, trend= 'add', damped_trend= True, seasonal= None, seasonal_periods= None).fit()
es_triple = ExponentialSmoothing(smoothing_df.O_3, trend= 'mul', seasonal= 'mul', seasonal_periods= 12).fit()
```

```python
plt.figure(figsize=(20, 5))

es_simple.fittedvalues.plot(color=[(230/255,159/255,0)], legend=True, title='Exponentially Smoothed and Forecast O_3')
es_simple.forecast(36).plot(color=[(230/255,159/255,0)], style='-..')
es_double.fittedvalues.plot(color=[(0,158/255,115/255)])
es_double.forecast(36).plot(color=[(0,158/255,115/255)], style='-..')
es_triple.fittedvalues.plot(color=[(0,114/255,178/255)])
es_triple.forecast(36).plot(color=[(0,114/255,178/255)], style='-..')

plt.legend(['ses', 'ses forecast', 'double exp', 'double exp forecast',
            'triple exp', 'triple exp forecast'])
plt.show()
```

```python




smoothing_df['simple_es'] = es_simple.fittedvalues
smoothing_df['double_es'] = es_double.fittedvalues
smoothing_df['triple_es'] = es_triple.fittedvalues
print('Simple Exponential Smoothing MSE: ', round(es_simple.sse/len(es_simple.fittedvalues), 2))
print('Double Exponential Smoothing MSE: ', round(es_double.sse/len(es_double.fittedvalues), 2))
print('Triple Exponential Smoothing MSE: ', round(es_triple.sse/len(es_triple.fittedvalues), 2))
```

### Hodrick-Prescott Filter (HPF)

El filtro Hodrick-Prescott es una tcnica de filtracin de datos para tratar los ciclos de negocios a partir de datos econmicos. Desafortunadamente nuestros datos en esta notebook no son un buen candidato para cmo usar el filtro de Hodrick-Prescott, pero lo he incluido porque es usado en los datos econmicos a largo plazo que a menudo tienen ciclos asociados con los cambios en el mercado.



Los ciclos de negocios son difciles de capturar y representan una barrera a la estacionaridad. Podemos usar el filtro Hodrick-Prescott para corregir eso. HPF intenta descomponer la serie de tiempos en componentes de tendencia y ciclo.




El parmetro clave de suavizacin que proporcionamos es lambda, 1600 es el valor sugerido para los datos anuales y valores ms grandes se pueden utilizar para datos ms voltiles.



Otros filtros que son populares para el anlisis econmico y los ciclos de descomposicin son:

* Filtro de Baxter-King

* Filtro de Christiano-Fitzgerald

```python
filter_df = pd.DataFrame(monthly_df.O_3)
O3_cycle, O3_trend = hpfilter(filter_df, lamb=129600)
filter_df['cycle'] = O3_cycle
filter_df['trend'] = O3_trend

filter_df.plot(figsize=(20, 5), title='O_3 Pollutant Plot of Cycle and Trend')
```
